## 2. K-means Clustering

```python
def dist2(x, y):
    return np.sum((x - y) ** 2)
def NMI(P):
    Hxy = scipy.stats.entropy(P.reshape(-1))
    Hx = scipy.stats.entropy(P.sum(axis = 1))
    Hy = scipy.stats.entropy(P.sum(axis = 0))
    I = Hx + Hy - Hxy # a famous equation
    return I / (Hx + Hy)
class Cluster:
    def __init__(self, node):
        self.avrg = node
        self.centre = None
        self.node_count = 0
        self.nodes = []
    def up_to_date(self):
        self.centre = self.avrg
        self.avrg = np.zeros(2)
        self.node_count = 0
        self.nodes = []
    def update(self, node):
        self.avrg = self.avrg * (self.node_count / (self.node_count + 1)) + node / (self.node_count + 1)
        self.node_count += 1
        self.nodes.append(node)
    def cal_average_dist(self, node):
        return np.mean(
            np.sqrt(
                np.sum(
                    (
                        np.array(self.nodes) 
                         - 
                        node.reshape(1, 2)
                    ) ** 2, 
                    axis = -1
                )
            )
        )
    
class Clusters:
    def __init__(self, data_points, K, max_iter):
        self.max_iter = max_iter
        self.K = K
        self.data = data_points
        self.clusters = [
            Cluster(data_points[index])
             for index in np.random.randint(0, data_points.shape[0], size = K)
        ]
    def compute_belong(self):
        self.MID = np.sqrt(
                        np.sum(
                              (
                                self.data.reshape(-1, 1, 2) 
                                 - 
                                np.array(
                                    [cl.centre for cl in self.clusters]
                                ).reshape(1, -1, 2)
                               ) ** 2, # diff ^ 2
                               axis = -1
                            )
                        ) # compute the pair-wise distance
        self.belongs = np.argmin(self.MID, axis = 1)
        self.MID[np.arange(self.data.shape[0]), self.belongs] = 1e9
        self.SilhoutteBelong = np.argmin(self.MID, axis = 1)
    def S(self, i):
        A = self.clusters[self.belongs[i]].cal_average_dist(self.data[i])
        B = self.clusters[self.SilhoutteBelong[i]].cal_average_dist(self.data[i])
        return (B - A) / max(A, B)
    def fit(self):
        for i in range(self.max_iter):
            # up-to-date the newest centre
            for cluster in self.clusters:
                cluster.up_to_date()
            # compute_belongs
            self.compute_belong()
            # update each cluster
            for j, Data in enumerate(self.data):
                self.clusters[self.belongs[j]].update(Data)
    def metric(self, type, extra_data = None):
        if type == 1: # within-cluster sum of squared distances
            ret = np.sum([ 
                        dist2(
                            self.clusters[self.belongs[j]].centre, 
                            self.data[j]
                        ) 
                        for j in range(self.data.shape[0])
                    ])
        elif type == 2:
            ret = np.mean([self.S(j) for j in range(self.data.shape[0])])
        elif type == 3:
            prob_matrix = np.zeros((10, self.K))
            for j, Data in enumerate(self.data):
                # print(extra_data[j], self.belongs[j])
                prob_matrix[int(extra_data[j])][self.belongs[j]] += 1
            prob_matrix /= prob_matrix.sum()
            ret = NMI(prob_matrix)
        return ret
```

```python
def kmeans(data, K, max_it, if_print = True):
    OBJ = Clusters(data[:, 2:], K, max_it)
    OBJ.fit()
    WC_SSD, SC, NMI = OBJ.metric(1), OBJ.metric(2), OBJ.metric(3, data[:, 1])
    if if_print:
        print(f"WC-SSD: {WC_SSD}\nSC: {SC}\nNMI: {NMI}\n")
    return WC_SSD, SC, NMI
```

### 2.2 Analysis

#### 2.2.1 Build 3 dataset

```python
data_set_1 = NP_data # org: NP_data
filter_2, filter_4, filter_6, filter_7 = NP_data[:, 1] == 2, NP_data[:, 1] == 4, NP_data[:, 1] == 6, NP_data[:, 1] == 7
data_set_2 = NP_data[filter_2 | filter_4 | filter_6 | filter_7]
data_set_3 = NP_data[filter_6 | filter_7]
```

#### 2.2.2 Build the Figure

```python
def Run_with_seed(SEED):
    np.random.seed(SEED)
    metric_1 = [kmeans(data_set_1, k, 50, False)[:2] for k in (2, 4, 8, 16, 32)]
    metric_2 = [kmeans(data_set_2, k, 50, False)[:2] for k in (2, 4, 8, 16, 32)]
    metric_3 = [kmeans(data_set_3, k, 50, False)[:2] for k in (2, 4, 8, 16, 32)]
    WC_SSD_1, SC_1 = zip(*metric_1)
    WC_SSD_2, SC_2 = zip(*metric_2)
    WC_SSD_3, SC_3 = zip(*metric_3)
    return WC_SSD_1, WC_SSD_2, WC_SSD_3, SC_1, SC_2, SC_3
```

```python
def Plot_(K_values, D1, D2, D3, name):
    plt.plot(K_values, D1, label = "dataset_1", marker='o')
    plt.plot(K_values, D2, label = "dataset_2", marker='o')
    plt.plot(K_values, D3, label = "dataset_3", marker='o')
    plt.xlabel("Value of k")
    plt.ylabel(name)
    plt.title(f"Effect of k on {name}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
WC_SSD_1, WC_SSD_2, WC_SSD_3, SC_1, SC_2, SC_3 = Run_with_seed(0)
KSET = [2, 4, 8, 16, 32]
Plot_(KSET, WC_SSD_1, WC_SSD_2, WC_SSD_3, "WC_SSD")
Plot_(KSET, SC_1, SC_2, SC_3, "SC")
```

- for dataset1, The best $K$ is $16$:
    1. The WC_SSD becomes slowly decreased after $K=16$, meaning it's already a point near the convergence.
    2. The SC in $K=16$ is largest, which means $K=16$ has the best distinction among all k.
- For similar reasons, I'll choose $K=4$ and $K=2$ for the second and third dataset.

```python
def report_mean_std(arr, name, which_ds, k):
    print(f"The mean = {np.mean(arr)}, std = {np.std(arr)} (of Metric {name} on dataset{which_ds}, AND K = {k}) ")

for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[0][idx], "WC_SSD", 1, k)
for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[1][idx], "WC_SSD", 2, k)
for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[2][idx], "WC_SSD", 3, k)
for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[3][idx], "SC", 1, k)
for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[4][idx], "SC", 2, k)
for idx, k in enumerate(KSET):
    report_mean_std(Total_metric_arr[5][idx], "SC", 3, k)
```

- We found that, if the $K$ is very different from the real clusters' number, it's sensitivity of inital points is high.

#### 2.2.4 NMI

```python
# K = (16, 4, 2) for dataset (1, 2, 3)
np.random.seed(0)
def work_on_NMI(dataset_, K, which_ds, max_it = 50):
    OBJ = Clusters(dataset_[:, 2:], K, max_it)
    OBJ.fit()
    NMI = OBJ.metric(3, dataset_[:, 1])

    print(f"NMI on dataset {which_ds} = {NMI} (Using the best-fit K = {K})")
    
    # plot
    selected = np.random.randint(0, dataset_.shape[0], size = 1000)
    plt.scatter(
        dataset_[selected, 2], 
        dataset_[selected, 3], 
        c = OBJ.belongs[selected], 
        cmap = 'viridis'
    )
    plt.colorbar()
    plt.title(f"Effect of Cluster on dataset {which_ds}")
    plt.show()

work_on_NMI(data_set_1, 16, 1)
work_on_NMI(data_set_2, 4, 2)
work_on_NMI(data_set_3, 2, 3)
```

- We observe that in the last two dataset, because $K=real\ clusters'\ number$, the cluster can be described as `nearly perfect`.
    - No matter in visilization or the NIM, we can get this conclusion.
    - The visilization is trivial.
    - Note that the two NIM values are near $0.5$, which means the relationship between real distribution and cluter distribution is really close!

## 3. Hierarchical Clusters

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

np.random.shuffle(NP_data)
final_dataset = np.array([NP_data[NP_data[:, 1] == u][:10] for u in range(10)]).reshape(-1, 4)

# Hierarchical CLUSTERs
linked = linkage(final_dataset, 'single')
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation = 'top', distance_sort = 'descending', show_leaf_counts = True)
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.show()
```

```python
# Hierarchical CLUSTERs
from sklearn.metrics import silhouette_score
linked = linkage(final_dataset, 'complete')
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation = 'top', distance_sort = 'descending', show_leaf_counts = True)
plt.title('Hierarchical Clustering Dendrogram (Complete Linkage)')
plt.show()
```

```python
# Hierarchical CLUSTERs
linked = linkage(final_dataset, 'average')
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation = 'top', distance_sort = 'descending', show_leaf_counts = True)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.show()
```

### 3.3

```python
def Plot_Hier(meth):
    linked = linkage(final_dataset, method = meth)
    
    ks = [2, 4, 8, 16, 32]
    wc_ssds = []
    scs = []
    
    for k in ks:
        clusters = fcluster(linked, k, criterion = 'maxclust')
        wc_ssd = sum(
            np.sum(
                (
                    final_dataset[clusters == i] - final_dataset[clusters == i].mean(axis = 0)
                ) ** 2
            ) 
            for i in range(1, k+1)
        )
        wc_ssds.append(wc_ssd)
        sc = silhouette_score(final_dataset, clusters)
        scs.append(sc)

    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(ks, wc_ssds, marker='o')
    plt.title(f'Within-Cluster Sum of Squares by K (method = {meth})')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('WC SSD')

    plt.subplot(1, 2, 2)
    plt.plot(ks, scs, marker='o')
    plt.title(f'Silhouette Coefficient by K (method = {meth})')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Silhouette Coefficient')
    
    plt.tight_layout()
    plt.show()
```

```python
Plot_Hier('single')
Plot_Hier('complete')
Plot_Hier('average')
```

```python
def Custom_NMI(meth):
    linked = linkage(final_dataset, method = meth)
    clusters = fcluster(linked, 32, criterion = 'maxclust')
    prob_matrix = np.zeros((10, 32))
    extra_data = final_dataset[:, 1]
    for j, Data in enumerate(final_dataset):
        prob_matrix[int(extra_data[j])][clusters[j]-1] += 1
    prob_matrix /= prob_matrix.sum()
    ret = NMI(prob_matrix)
    return ret

print(f"NMI of method single = {Custom_NMI('single')}")
print(f"NMI of method complete = {Custom_NMI('complete')}")
print(f"NMI of method average = {Custom_NMI('average')}")
```
